```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
data = pd.read_excel('population_by_country_2020.xlsx')
data.head()
```

```python
fig = px.choropleth(data, locations='Country (or dependency)',
                                    locationmode="country names", color='Population (2020)', scope="world")

# Display the plotly graph as a static image
fig.show("png")

# Output the plot as html and save as population.html (interactive html file)
# To view the file go to http://htmlpreview.github.io/?https://github.com/rosemonga91/population-by-country-2020/blob/main/population.html
pio.write_html(fig, file='population.html')
```
